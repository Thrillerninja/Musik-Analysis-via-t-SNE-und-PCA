```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definition der Sigmoid-Funktion und ihrer Ableitung
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

A few common sigmoud functions

```python
from scipy.special import erf

# Hyperbolic Tangent
def tanh(x):
    return np.tanh(x)

# Smoothstep function
def smoothstep(x):
    return np.clip(x, 0, 1)

# Error function
def erf_function(x):
    return erf(x)

# Generate data
x_large = np.linspace(-10, 10, 1000)
sigmoid_values = sigmoid(x_large)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x_large, sigmoid_values, label=r"Logistic: $\frac{1}{1 + e^{-x}}$", color="blue")
plt.plot(x_large, tanh(x_large), label=r"$\tanh(x)$", color="green", alpha=0.5)
plt.plot(x_large, erf_function(x_large), label=r"$\frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} dt$", color="purple", alpha=0.5)
plt.plot(x_large, np.arctan(x_large), label=r"$\arctan(x)$", color="orange", alpha=0.5)

# Add titles and labels
plt.xlabel("x")
plt.ylabel("Function values")

# Add grid
plt.grid(True)

# Add legend 
plt.legend()

# Save file to be used in latex
plt.savefig("sigmoid_set.png")

# Show plot
plt.show()
```

Plot Standard Sigmoid Function

```python
# Generate data
x_large = np.linspace(-10, 10, 1000)
sigmoid_values = sigmoid(x_large)

# Plot
plt.figure(figsize=(8, 3))
plt.plot(x_large, sigmoid_values, label="Sigmoid", color="blue")

# Add titles and labels
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")

# Add grid
plt.grid(True)

# save file to be used in latex
plt.savefig("sigmoid.png")

# Add legend and show plot
plt.show()
```

& Derivative

```python
# Wertebereich fr die Visualisierung
x = np.linspace(-10, 10, 1000)

# Plot der Aktivierungsfunktionen
plt.figure(figsize=(8, 4))

# Sigmoid
plt.plot(x, sigmoid_derivative(x), label="Sigmoid Ableitung", linestyle="--")
plt.xlabel("x")
plt.ylabel("Sigmoid'(x)")

# save file to be used in latex
plt.savefig("sigmoid_abl.png")

# Add legend and show plot
plt.show()
```
